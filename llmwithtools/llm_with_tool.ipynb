```python
# Arxiv Tool

arxivapi = ArxivAPIWrapper(top_k_results=2, ARXIV_MAX_QUERY_LENGTH=500)
arxiv = ArxivQueryRun(arxiv_api_wrapper=arxivapi)
```

```python
# Wikipedia Tool

wikiapi = WikipediaAPIWrapper(top_k_results=1, doc_content_chars_max=500)
wiki = WikipediaQueryRun(api_wrapper=wikiapi)
```
